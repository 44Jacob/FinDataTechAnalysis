```python
import yfinance as yf
import pandas as pd

# Define the ticker symbols for the companies of interest
ticker_symbols = ['MSFT', 'CRM', 'CRWD', 'ZM', 'SHOP']

# Create an empty DataFrame to store summary data
summary_data = pd.DataFrame(columns=['Ticker', 'Latest YoY Revenue Growth', 'Latest R&D Expense % of Revenue'])

for ticker in ticker_symbols:
    company = yf.Ticker(ticker)
    
    # Try to fetch annual income statement, if available
    income_statement = company.financials.T.dropna(how='all')
    if 'Research Development' in income_statement.columns and 'Total Revenue' in income_statement.columns:
        # Calculate R&D Expenses as a Percentage of Revenue
        rd_expenses = income_statement['Research Development'].fillna(0)
        revenue = income_statement['Total Revenue']
        
        # Ensure there's enough data to calculate YoY growth and percentages
        if len(revenue) > 1 and revenue.iloc[0] != 0:
            yoy_revenue_growth = ((revenue.iloc[0] - revenue.iloc[1]) / revenue.iloc[1]) * 100
            rd_expense_percentage = (rd_expenses.iloc[0] / revenue.iloc[0]) * 100
            
            # Append the calculated data to the summary DataFrame
            summary_data = summary_data.append({
                'Ticker': ticker,
                'Latest YoY Revenue Growth': f"{yoy_revenue_growth:.2f}%",
                'Latest R&D Expense % of Revenue': f"{rd_expense_percentage:.2f}%"
            }, ignore_index=True)
        else:
            print(f"Insufficient data for {ticker} to calculate YoY Revenue Growth or R&D Expense %.")
    else:
        print(f"R&D Expense data not available for {ticker}.")

print(summary_data)
```

```python
import yfinance as yf
import pandas as pd

def get_data_safe(df, key):
    try:
        return df.loc[key].iloc[-1]  # Adjusted to fetch the most recent data available
    except KeyError:
        return None

# List of ticker symbols for the companies of interest
tickers = ['MSFT', 'CRM', 'CRWD', 'ZM', 'SHOP']

for ticker in tickers:
    company = yf.Ticker(ticker)
    
    # Fetch financial statements and market data
    income_statement = company.financials.T
    balance_sheet = company.balance_sheet.T
    cashflow_statement = company.cashflow.T
    info = company.info
    
    # Market Cap
    market_cap = info.get('marketCap', 'N/A')

    # Profit Margin (Net Income / Total Revenue)
    net_income = get_data_safe(income_statement, 'Net Income')
    total_revenue = get_data_safe(income_statement, 'Total Revenue')
    profit_margin = (net_income / total_revenue) if net_income and total_revenue else 'N/A'
    
    # ROE (Net Income / Shareholder's Equity)
    shareholders_equity = get_data_safe(balance_sheet, 'Total Stockholder Equity')
    roe = (net_income / shareholders_equity) if net_income and shareholders_equity else 'N/A'
    
    # ROA (Net Income / Total Assets)
    total_assets = get_data_safe(balance_sheet, 'Total Assets')
    roa = (net_income / total_assets) if net_income and total_assets else 'N/A'
    
    # Current Ratio (Current Assets / Current Liabilities)
    current_assets = get_data_safe(balance_sheet, 'Total Current Assets')
    current_liabilities = get_data_safe(balance_sheet, 'Total Current Liabilities')
    current_ratio = (current_assets / current_liabilities) if current_assets and current_liabilities else 'N/A'
    
    # Quick Ratio ((Current Assets - Inventory) / Current Liabilities)
    inventory = get_data_safe(balance_sheet, 'Inventory') or 0
    quick_ratio = ((current_assets - inventory) / current_liabilities) if current_assets and current_liabilities else 'N/A'
    
    # Debt-to-Equity Ratio (Total Liabilities / Shareholders' Equity)
    total_liabilities = get_data_safe(balance_sheet, 'Total Liab')
    debt_to_equity = (total_liabilities / shareholders_equity) if total_liabilities and shareholders_equity else 'N/A'
    
    print(f"Ticker: {ticker}")
    print(f"Profit Margin: {profit_margin}")
    print(f"ROE: {roe}")
    print(f"ROA: {roa}")
    print(f"Current Ratio: {current_ratio}")
    print(f"Quick Ratio: {quick_ratio}")
    print(f"Debt-to-Equity Ratio: {debt_to_equity}")
    print(f"Market Cap: {market_cap}\n")
```
